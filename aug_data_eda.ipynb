```python
import pandas as pd
import numpy as np

import re
```

```python
data = pd.read_csv('../Data/training_set_rel3.tsv', sep='\t', encoding='mac_roman', index_col='essay_id')
data
```

```python
data = data[data['essay_set'].isin((1, 2, 7, 8))]
data
```

```python
data.columns
```

```python
data = data[['essay_set', 'essay', 'domain2_score']]
data
```

```python
unnamers_aug = set()
for _, row in data.iterrows():
    unnamers_aug |= set(re.findall(r'@[A-Z]+[1-9]+', row['essay']))
unnamers_aug
```

```python
aug_data = data
```

```python
orig_data = pd.read_csv('../Data/train.csv')
orig_data
```

```python
unnamers_orig = set()
for _, row in orig_data.iterrows():
    unnamers_orig |= set(re.findall(r'[a-zA-Z]+\_[a-zA-Z]+', row['full_text']))
unnamers_orig
```

```python
unnaming_mapping = {
    '@CAPS': 'Generic_Name',
    '@CITY': 'Generic_City',
    '@DATE': '', # rule based, replace with random date
    '@DR': 'Generic_Name',
    '@EMAIL': '',
    '@LOCATION': 'LOCATION_NAME',
    '@MONEY': 'OTHER_NAME',
    '@NUM': '', # random num
    '@ORGANIZATION': 'OTHER_NAME',
    '@PERCENT': '', #random num
    '@PERSON': 'Generic_Name',
    '@STATE': 'LOCATION_NAME',
    '@TIME': '',
}
```

```python
for old_name, new_name in unnaming_mapping.items():
    aug_data.loc[:, 'essay'] = aug_data['essay'].str.replace(fr'{old_name}[1-9]+', new_name, regex=True)
aug_data
```

```python
aug_data_sample = aug_data.groupby(['essay_set']).sample(250, random_state=42).rename({'essay': 'full_text'}, axis=1)['full_text']
aug_data_sample
```

```python
aug_data_sample.to_csv('../Data/aug_data_sample.csv', index=False)
```

```python
aug_data[['essay']].rename({'essay': 'full_text'}, axis=1).to_csv('../Data/aug_data.csv', index=False)
```
